```python
df_car.rename(columns={df_car.columns[0]:"년도", 
                      df_car.columns[1]:"자치구", 
                      df_car.columns[2]:"합계", 
                      df_car.columns[3]:"승용차", 
                      df_car.columns[4]:"승합차", 
                      df_car.columns[5]:"화물차", 
                      df_car.columns[6]:"특수차"}, 
              inplace=True)

df_car.head()
```

```python
df_par.rename(columns={df_par.columns[0]:"년도", 
                      df_par.columns[1]:"자치구", 
                      df_par.columns[2]:"합계", 
                      df_par.columns[3]:"노상", 
                      df_par.columns[4]:"노외", 
                      df_par.columns[5]:"건축물부설"},
             inplace=True)
df_par.head()
```

```python
df_par
```

```python
# 3. 승용차 데이터의 불필요한 행을 삭제합니다.
```

```python
df_car.drop([0],inplace=True) # 자동차 서울시 제거
df_car.drop([26],inplace=True) # 자동차, 기타 제거
df_par.drop([0],inplace=True) # 주차장 서울시 제거
```

```python
df_car
```

```python
df_par
```

```python
df_car.info()
```

```python
# 컬럼 데이터 형식 변환

df_car['합계'] = df_car['합계'].astype(int)
```

```python
df_car.info()
```

```python
# df_car에 승용차비율, 승합차비율, 화물차비율, 계산필드 추가

df_car['승용차비율']=df_car['승용차']/df_car['합계']*100
df_car['승합차비율']=df_car['승합차']/df_car['합계']*100
df_car['화물차비율']=df_car['화물차']/df_car['합계']*100
df_car.head(10)
```

```python
# 노상비율, 노외비율, 건축물비율 계산필드 추가

df_par['노상비율']=df_par['노상']/df_par['합계']*100
df_par['노외비율']=df_par['노외']/df_par['합계']*100
df_par['건축물비율']=df_par['건축물부설']/df_par['합계']*100
df_par.head(10)
```

```python
df_car1 = df_car
```

```python
df_car1.head()
```

```python
df_car1=df_car.copy()
df_par1=df_par.copy()
```

```python
del df_car1["승용차"]
del df_car1["승합차"]
del df_car1["화물차"]
del df_car1["특수차"]
df_car1.head()
```

```python
del df_par1["노상"]
del df_par1["노외"]
del df_par1["건축물부설"]
df_par1.head()
```

```python
data_seoul=pd.merge(df_car1, df_par1, 
                   on="자치구")
data_seoul.head()
```

```python
data_seoul.rename(columns={data_seoul.columns[0]:"년도", 
                          data_seoul.columns[2]:"차량합계", 
                          data_seoul.columns[7]:"주차장합계"},
                 inplace=True)
data_seoul.head()
```

```python
del data_seoul["년도_y"]
data_seoul.head()
```

```python
data_seoul.set_index("자치구", inplace=True)
data_seoul.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 맥북에서 plot에 한글 삽입
```

```python
ax = data_seoul[['차량합계','주차장합계']].plot(kind="bar", 
                                      title="주차장", 
                                      figsize=(12, 4), 
                                      legend=True, 
                                      fontsize=12)
ax.set_xlabel("자치구", fontsize=12)
ax.set_ylabel("차량/주차장", fontsize=12)
ax.legend(["차량", "주차장"], fontsize=12)
plt.show()
```
